 # Visual Analysis of the Dots Dataset

## import  libarires

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

###  load dataset

```python
# Step 1: Load the dataset
dots = sns.load_dataset("dots")
dots
```

#### Overview of the dataset

```python
print("Dataset Overview:")
print(dots.head())
```

```python
# Explore dataset information
print("\nDataset Info:")
print(dots.info())
```

```python
# Describe numerical features
print("\nDataset Description:")
print(dots.describe())
```

#### Data Cleaning 

```python
# Check for missing values
print("\nChecking for missing values:")
print(dots.isnull().sum())
```

```python
# Scatter plot to show the relationship between time and firing rate
plt.figure(figsize=(10, 6))
sns.scatterplot(data=dots, x="time", y="firing_rate", hue="coherence", style="choice")
plt.title("Firing Rate vs Time (Colored by Coherence and Styled by Choice)")
plt.xlabel("Time")
plt.ylabel("Firing Rate")
plt.legend(title="Coherence and Choice")
plt.show()
```

```python
# Line plot to show trends over time
plt.figure(figsize=(10, 6))
sns.lineplot(data=dots, x="time", y="firing_rate", hue="coherence", style="choice", markers=True)
plt.title("Trends in Firing Rate vs Time by Coherence and Choice")
plt.xlabel("Time")
plt.ylabel("Firing Rate")
plt.legend(title="Coherence and Choice")
plt.show()
```

```python
# Visualize firing rate distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=dots, x="firing_rate", hue="coherence", kde=True)
plt.title("Distribution of Firing Rate by Coherence")
plt.xlabel("Firing Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pairplot to explore multiple relationships
sns.pairplot(data=dots, hue="coherence", markers=["o", "s", "D"])
plt.suptitle("Pairplot of the Dots Dataset", y=1.02)
plt.show()
```

## Conclusions

```python
print("\nConclusions:")
print("- The dataset provides insights into how firing rates vary over time depending on coherence and choice.")
print("- The scatter and line plots reveal clear trends, with higher coherence generally leading to higher firing rates.")
print("- The distribution plot shows that the firing rate is concentrated around certain values for different levels of coherence.")
```
